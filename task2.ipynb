```python


class TrapShape:
    def __init__(self, top, bottom, height):
        self.top = top
        self.bottom = bottom
        self.height = height

    def area(self):
        return 0.5 * (self.top + self.bottom) * self.height


class ParaShape:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return self.base * self.height

data_list = [
    (5, 7, 4, 6, 3),
    (8, 10, 5, 7, 4),
    (4, 6, 3, 5, 2)
]

count = 1

for item in data_list:
    t1, t2, h1, b1, h2 = item

    shape1 = TrapShape(t1, t2, h1)
    shape2 = ParaShape(b1, h2)

    area_store = {
        "Trapezoid": shape1.area(),
        "Parallelogram": shape2.area()
    }

    print("\nSet", count)
    print("Areas:", area_store)

    big_shape = max(area_store, key=area_store.get)

    print("Maximum Area Shape:", big_shape)
    print("Area Value:", area_store[big_shape])

    count += 1
```
